```python
%matplotlib inline

from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt

# Initialize A Simple Linear Regression Model
reg = linear_model.LinearRegression()

f = open("data/BrainVsBodyWeight.tsv") # Read the data from File
f.readline() # Skip the First Line
tdata = np.loadtxt(f) # load the Data to a temporary Location
data = np.array(sorted(tdata, key=lambda x:x[1])) # Sort the data by X for a better Plot
x=list(map((lambda t:[t]),data[:,1]))
y = data[:,2]

reg.fit(x,y)

print("Coeffecients : ",reg.coef_)
print("Intercept : ",reg.intercept_)
print("Output for 3.385 : ",reg.predict(np.array([3.385]).reshape(-1,1)))
print("Score : ",reg.score(x,y))

z = list(map((lambda t:reg.predict(np.array([t]).reshape(-1,1))[0]),x))

plt.scatter(x,y,c=[1,0,0])
plt.scatter(x,z,c=[0,1,0])
plt.plot(x,y,x,z)
```
